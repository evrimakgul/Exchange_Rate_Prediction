# 1- Exchange Rates

```python
import numpy as np
import pandas as pd
```

```python
# we read our dataframe and assign the time column as the index values of the dataframe
forex = pd.read_csv("/home/evrim/Desktop/deneme/DATA/forex.csv", index_col='Date', parse_dates=True)
forex.head()
```

# 4- Libor Rates (Interest Rates for GBP and USD)

```python
liborGBP = pd.read_csv("/home/evrim/Desktop/deneme/DATA/libor_GBP.csv", index_col='Date', parse_dates=True, squeeze=True)
liborGBP
```

```python
liborUSD = pd.read_csv("/home/evrim/Desktop/deneme/DATA/libor_USD.csv", index_col='Date', parse_dates=True, squeeze=True)
liborUSD
```

```python
libor = pd.concat([liborGBP, liborUSD], axis=1)
libor[:11]
```

```python
# Some of the values are in "."
# we need to change them and convert the series to numeric
libor = libor.apply(pd.to_numeric, errors='coerce')
libor.info()
```

```python
merged3 = pd.merge(merged2, libor, left_index=True, right_index=True, how='outer')
merged3
```

```python
#interpolation needs continuous datetime
# or numerical objects for index values
merged3.reset_index(level=0, inplace=True)
merged3
```

```python
merged3.iloc[:, -6:] = merged3.iloc[:, -6:].interpolate()
merged3
```

```python
# drop rows where there is no data for GBP (or any forex)
merged3 = merged3[merged3.GBP.notna()]
merged3
```

```python
# there is not enough data for libor series at the beginning
# I drop the beginning rows of the whole dataframe
merged3 = merged3[merged3.liborGBP.notna()]
merged3
```
